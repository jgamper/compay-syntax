```python
from modules.openslideplus import OpenSlidePlus
from modules.annotation import Annotation
import modules.utils as ut
from sampler import Sampler

import numpy as np
import matplotlib.pyplot as plt
import glob
import os
%matplotlib inline
```

```python
def show_PIL(pil_im):
    plt.figure(figsize=(6,6))
    plt.imshow(np.asarray(pil_im),cmap='gray')
```

```python
mac = False
if mac:
    data_dir = '/Users/peterb/Dropbox/SharedMore/WSI_sampler'
else:
    data_dir = '/home/peter/Dropbox/SharedMore/WSI_sampler'

files = glob.glob(os.path.join(data_dir, '*.tif'))
print(files)

tm_dir = './tissue_masks'
annotation_dir = os.path.join(data_dir, 'annotation')
```

```python
file = files[1]
```

```python
osp = OpenSlidePlus(file, level0=40)
patch = osp.get_patch(w=45000, h=30000, mag=10, size=1000)
show_PIL(patch)
```

```python
annotation_file = os.path.join(annotation_dir, osp.ID + '_mask.tif')
ann = Annotation(annotation_file, reference_wsi=osp)
patch = ann.get_patch(w_ref=45000, h_ref=30000, mag=10, size=1000)
show_PIL(patch)
```

```python
sampler = Sampler(file, level0=40, tissue_mask_dir=tm_dir, annotation_dir=annotation_dir)
sampler.prepare_sampling(magnification=20, patchsize=240)
```

```python
sampler.sample_patches(max_per_class=100, savedir=os.getcwd())
```
